```python
#I want to give names that are more easily understood, from best to worse cut

data['cut'] = data['cut'].replace(['Ideal'],'1-Ideal')
data['cut'] = data['cut'].replace(['Premium'],'2-Premium')
data['cut'] = data['cut'].replace(['Very Good'],'3-Very Good')
data['cut'] = data['cut'].replace(['Good'],'4-Good')
data['cut'] = data['cut'].replace(['Fair'],'5-Fair')
```

```python
data.head()
```

```python
#I want to give names that are more easily understood, from best to worse color: 'D', 'E', 'F', 'G', 'H', 'I', 'J'

data['color'] = data['color'].replace(['D'],'1-D')
data['color'] = data['color'].replace(['E'],'2-E')
data['color'] = data['color'].replace(['F'],'3-F')
data['color'] = data['color'].replace(['G'],'4-G')
data['color'] = data['color'].replace(['H'],'5-H')
data['color'] = data['color'].replace(['I'],'6-I')
data['color'] = data['color'].replace(['J'],'7-J')
```

```python
data.head()
```

```python
data['color'].unique()
```

```python
data['clarity'].unique()
```

```python
data['clarity'] = data['clarity'].replace(['IF'],'1-IF')
data['clarity'] = data['clarity'].replace(['VVS1'],'2-VVS1')
data['clarity'] = data['clarity'].replace(['VVS2'],'3-VVS2')
data['clarity'] = data['clarity'].replace(['VS1'],'4-VS1')
data['clarity'] = data['clarity'].replace(['VS2'],'5-VS2')
data['clarity'] = data['clarity'].replace(['SI1'],'6-SI1')
data['clarity'] = data['clarity'].replace(['SI2'],'7-SI2')
data['clarity'] = data['clarity'].replace(['I1'],'8-I1')
```
